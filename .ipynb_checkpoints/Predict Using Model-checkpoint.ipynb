```python
def prepare(path):
    print("Prepare ",path)
#     IMG_SIZE = 50
    img_array = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
#     return img_array

#     plt.imshow(img_array,cmap = 'gray')
#     plt.show()
    new_array = cv2.resize(img_array,(120,100))
    new_array = new_array.reshape(120,100,1)
    return new_array

def importAll(path):
    all_images = []
    for subfolder in os.listdir(path):
        folder = os.path.join(path,subfolder)
        for file in os.listdir(folder):
            f_proc = os.path.join(folder,file)
            x = prepare(f_proc)
            all_images.append(x)
    return all_images
```
